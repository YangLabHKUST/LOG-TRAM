## Download data

We illustrate the usage of LOG-TRAM by applying it to the GWAS summary statistics of SBP from BBJ and UKBB with 1 Mbp non-overlapping sliding windows as local regions. The GWAS datasets and LDscores files involved in the following example are availabel from [here](https://www.dropbox.com/sh/9asugdlu1lbal8o/AAB0martsgaBoR8B4hq2pc25a?dl=0)

## Run LOG-TRAM

Once the input files are formatted, LOG-TRAM will automatically preprocess the datasets, including SNPs overlapping and minor allele matching. It takes 8 mins to run the following meta-analysis for the whole genome (computing environment: 20 CPU cores of Intel(R) Xeon(R) Gold 6230N CPU @ 2.30GHz processor, 1TB of memory, and a 22 TB solid-state disk). 

```python
python <install path>/src/LOG-TRAM.py \
        --out SBP_meta \
        --sumstats-popu1 SBP_harmonized_pop1_UKB.txt,SBP_UKB \
        --sumstats-popu2 SBP_harmonized_pop2_BBJ.txt,SBP_BBJ \
        --ldscores ./LDscoresEUR-EAS/ldsc_annot_EUR_EAS_1mb_TGP_hm3_chr@_std
```

### Manhatton plot

```python
# identify lead SNPs from GWAS summary statistics
def add_locus(gwas):
    gwas_sig = gwas.loc[gwas['P']<=5e-8].reset_index(drop=True)
    loci = []
    c = 0
    bp = -1
    l = -1
    for row in gwas_sig.iterrows():
        if c!=row[1]['CHR'] or row[1]['BP']-bp>1000000:
            c = row[1]['CHR']
            bp = row[1]['BP']
            l += 1
        loci.append(l)
    gwas_sig['loci'] = loci
    gwas_sig = gwas_sig.loc[gwas_sig.groupby('loci')['P'].idxmin()].reset_index(drop=True)
    return gwas_sig
```

```python
# maximum p-vaule = 1e-30
threshold = 1e-30
eas_gwas.loc[eas_gwas['P']<threshold,'P'] = threshold
eas_meta.loc[eas_meta['P']<threshold,'P'] = threshold
```

```python
eas_gwas_sig = add_locus(eas_gwas)
eas_meta_sig = add_locus(eas_meta)
```

## Effective sample size

```python
eas_gwas = pd.read_csv('SBP_harmonized_pop2_BBJ.txt',sep='\t')
eas_meta = pd.read_csv('SBP_meta_TRAM_pop2_SBP_BBJ.txt',sep='\t')
```

```python
# intercept_gwas: LDSC regression intercept for Original EAS GWAS summary statistics
# intercept_meta: LDSC regression intercept for LOG-TRAM EAS meta-analysis association statistics
# Both of them are available in output log

intercept_meta, intercept_gwas = 1.0221411, 1.05924058

Neff_f = ((eas_meta['Z']**2).mean() - intercept_meta)/((eas_gwas['Z']**2).mean() - intercept_gwas)
print('Original EAS GWAS sample size: {},\nLOG-TRAM EAS effective sample size: {}'.format(
    eas_gwas['N'].values.mean(), Neff_f*eas_gwas['N'].values.mean()))
```
